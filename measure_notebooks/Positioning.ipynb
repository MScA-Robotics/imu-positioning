```python
### finding the turn
y=[]
x=[]
euler_x_delta=[]
y_delta_list=[]
x_delta_list=[]
xy_delta=[]
for i in range(len(df_new['euler_x'])):
    if (i==0):
        delta=0
        y_delta=0
        x_delta=0
    else:
        delta=df_new['euler_x'][i]-df_new['euler_x'][i-1]
        y_delta=df_new['left_enc'][i]-df_new['left_enc'][i-1]
        x_delta=df_new['right_enc'][i]-df_new['right_enc'][i-1]
    try:
        
        if( delta<-300):
            delta=delta+360
        if( delta>300):
            delta=delta-360
        else:
            delta=delta
        euler_x_delta.append(delta)
    except:
        euler_x_delta.append(0)
        
    y_delta_list.append(y_delta)
    x_delta_list.append(x_delta)
    xy_delta.append(y_delta-x_delta)
    y.append(math.sin(df_new['euler_x'][i]*0.0174533)*y_delta)
    x.append(math.cos(df_new['euler_x'][i]*0.0174533)*x_delta)

df_new['euler_x_delta']=euler_x_delta
df_new['y_delta_list']=y_delta_list
df_new['x_delta_list']=x_delta_list
df_new['xy_delta']=xy_delta
df_new['y']=y
df_new['x']=x
```
